```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/netflix_titles.csv')

# Show first 5 rows
df.head()
```

```python
df.info()
df.shape
df.columns
```

```python
df.isnull().sum()
```

```python
# Remove rows where 'title' is missing (important column)
df.dropna(subset=['title'], inplace=True)

# Fill missing 'director' or 'cast' with 'Unknown'
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
# Convert all string data to lowercase for consistency
df['type'] = df['type'].str.lower()
df['country'] = df['country'].str.lower()
df['listed_in'] = df['listed_in'].str.lower()
```

```python
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

# Extract year from date_added (optional)
df['year_added'] = df['date_added'].dt.year
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.info()
df.head()
```

```python
df.to_csv('cleaned_netflix.csv', index=False)
```

## Summary of Data Cleaning

- Removed 10 duplicate rows using `drop_duplicates()`
- Filled missing values in `country` with "Unknown"
- Standardized `date_added` to `dd-mm-yyyy` format
- Renamed columns to lowercase
- Checked and fixed data types
- Saved the cleaned dataset
